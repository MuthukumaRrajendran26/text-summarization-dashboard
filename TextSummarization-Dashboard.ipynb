```python
from transformers import pipeline

summarizer = pipeline("summarization")
```

```python
# serverDownload = Flask(__name__)
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP], title='NSERC Project')


#app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

main = dbc.Container([ 
    html.Div([
        html.Hr(),
        
        dbc.Row([html.H6("Copy and paste the text that you want to summarize in the below area")]),
        
        html.Hr(),

        dbc.Row([
            dbc.Textarea(id="textInput", className="mb-3", placeholder="Enter text"),
        ]),

        html.Hr(),

        dbc.Row([
            dbc.Button("Summarize", id = 'Summarize', color="primary", className="mr-1"),
        ]),

        html.Hr(),
        
        dbc.Row([html.H6("Summary for the above text")]),
        
        html.Hr(),
        
        dbc.Row([
            dbc.Textarea(id="sumOut", className="mb-3")
        ]),
        
        html.Hr()
    ]),
])

app.layout = main


@app.callback(  # See documentation or tutorial to see how to use this
    Output('sumOut', 'value'),
    Input('Summarize', 'n_clicks'),
    State('textInput', 'value'))
def update_linePlot(n_clicks, value):
    if n_clicks == 1 and value != '' or value != None:
        summary = summarizer(value, max_length=130, min_length=30, do_sample=False)
        result = summary[0]["summary_text"]
        return result
    
app.run_server(mode='jupyterlab', dev_tools_ui=True, debug=True, dev_tools_hot_reload =True, threaded=True)
```
